```python
df = df.rename(columns={"Unnamed: 0":"Date"})
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df.Date, dayfirst=True)
df2['Dates'] = pd.to_datetime(df2.Dates, dayfirst=True)
```

```python
#Top States that consumed maximum power
mean_temperature = df.mean().sort_values(ascending=False).reset_index().rename(columns = {"index": "state", 0 : "avg_consumption"})
mean_temperature
```

```python
#Top 10 states with least consumption
mean_temperature_small = df.mean().sort_values(ascending=True).reset_index().rename(columns = {"index": "state", 0 : "avg_consumption"})
mean_temperature_small
```
